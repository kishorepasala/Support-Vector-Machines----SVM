# data preprocessing

```python
le = LabelEncoder()
df["month"] = le.fit_transform(df["month"])
df["day"] = le.fit_transform(df["day"])
df["size_category"] = le.fit_transform(df["size_category"])
```

```python
df.head()
```

```python
df.drop(['month','day'], axis = 1, inplace = True)
```

```python
std = StandardScaler()
x = df.iloc[:,:9]
df.iloc[:,:9] = std.fit_transform(x)
```

```python
df
```

```python
y = df['size_category']
df.drop(['size_category'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size = 0.25, random_state = 22)
```

```python
svc = SVC()
param_grid = [{'kernel':['rbf'], 'gamma': [0.2,0.5,1,5,10,20], 'C': [1,2,4,8,10,20,40,50,80,100]}]
gsv = GridSearchCV(svc, param_grid, cv = 10 )
gsv.fit(x_train, y_train)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

```python
svc = SVC(C = 1, gamma = 0.2)
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
classification_report(y_test, y_pred)
```
